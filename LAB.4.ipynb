```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid function with parameters for weight (w) and bias (b)
def sigmoid(x, w=1, b=1):
    return 1 / (1 + np.exp(-(w * x + b)))
```

```python
# Generate x values over a range
x = np.linspace(-10, 10, 400)

# Define different values for weight (steepness) and bias (shift)
w_values = [1.5, 2, 3]   # Steepness of the curve
b_values = [0, 1, 2]     # Horizontal shift

# Define colors for plots
colors = ['b', 'orange', 'g']
```

```python
plt.figure(figsize=(12, 5))  # Create a 1x2 subplot

# Left plot: Effect of weight (w) with fixed bias b = 2
plt.subplot(1, 2, 1)
for i, w in enumerate(w_values):
    plt.plot(x, sigmoid(x, w=w, b=2), color=colors[i], linestyle='--', label=f'w = {w}, b = 2')
plt.title('Effect of w (Steepness) on Sigmoid')
plt.xlabel('x')
plt.ylabel('σ(x)')
plt.legend()
plt.grid(True)
```

```python
# Right plot: Effect of bias (b) with fixed steepness w = 2
plt.subplot(1, 2, 2)
for i, b in enumerate(b_values):
    plt.plot(x, sigmoid(x, w=2, b=b), color=colors[i], linestyle='-', label=f'w = 2, b = {b}')
plt.title('Effect of b (Position) on Sigmoid')
plt.xlabel('x')
plt.ylabel('σ(x)')
plt.legend()
plt.grid(True)

# Show both subplots
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid function with adjustable weight and bias
def sigmoid(x, w=1, b=0):
    return 1 / (1 + np.exp(-(w * x + b)))
```

```python
# Generate x values from -10 to 10
x = np.linspace(-10, 10, 400)

# Values for bias (position shift) and fixed weight (steepness)
w_values = [1.5, 2, 3]      # (not used in this plot but defined)
b_values = [0, 3, 4]        # Bias shift values

# Colors for curves
colors = ['b', 'orange', 'g']
```

```python
# Set figure size
plt.figure(figsize=(10, 4))

# Subplot for effect of bias
plt.subplot(1, 2, 1)
for i, b in enumerate(b_values):
    plt.plot(x, sigmoid(x, w=1, b=b), color=colors[i], linestyle='dotted', label=f'w = 1, b = {b}')

# Plot formatting
plt.title('Effect of b (Position) on Sigmoid')
plt.xlabel('x')
plt.ylabel('σ(x)')
plt.legend()
plt.grid(True)
```

```python
# Adjust layout and show
plt.tight_layout()
plt.show()
```
